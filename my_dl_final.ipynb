```python
from keras.models import Sequential
from keras.layers import Embedding, Bidirectional, LSTM, Dense, SpatialDropout1D
from keras.optimizers import Adam
from keras.callbacks import EarlyStopping


# Set learning rate
learning_rate = 0.001

# Instantiate the Adam optimizer with the specified learning rate
adam_optimizer = Adam(learning_rate=learning_rate)
max_feat = 2000
# Define the model
embed_dim = 64
bilstm = Sequential()
bilstm.add(Embedding(max_feat, embed_dim, input_length=None))
bilstm.add(SpatialDropout1D(0.2))
bilstm.add(Bidirectional(LSTM(128, dropout=0.2, recurrent_dropout=0.2)))
bilstm.add(Dense(5, activation='softmax'))

# Compile the model with the Adam optimizer
bilstm.compile(loss='categorical_crossentropy', optimizer=adam_optimizer, metrics=['acc'])
bilstm.summary()

```

```python
batch_size = 64
epochs  = 10
history1 = bilstm.fit(xtrain_pad, y_train_one_hot, class_weight=class_weight, validation_split=0.1, batch_size = batch_size, epochs = epochs)
```

```python
from sklearn.metrics import classification_report, accuracy_score, precision_score, f1_score, recall_score

# Evaluate the model on the test set
y_pred = bilstm.predict(xtest_pad)

# Convert one-hot encoded predicted labels to class labels
y_pred_classes = y_pred.argmax(axis=-1)

# Calculate evaluation metrics on the updated model
accuracy_bilstm = accuracy_score(ytest, y_pred_classes)
precision_bilstm = precision_score(ytest, y_pred_classes, average='weighted')
recall_bilstm = recall_score(ytest, y_pred_classes, average='weighted')
f1_bilstm = f1_score(ytest, y_pred_classes, average='weighted')

# Print or use the updated evaluation metrics
print("BiLSTM Evaluation Metrics:")
print(f"BiLSTM Accuracy: {accuracy_bilstm}")
print(f"BiLSTM Precision: {precision_bilstm}")
print(f"BiLSTM Recall: {recall_bilstm}")
print(f"BiLSTM F1 Score: {f1_bilstm}")
print('-----------------------------------------------------')
# Print the classification report
class_rep_bilstm = classification_report(ytest, y_pred_classes)
print("Classification Report of BiLSTM:\n", class_rep_bilstm)
```

```python
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report

# Assuming df is your DataFrame with 'Description' and 'Label' columns
# Split the data into training and testing sets
train_data, test_data, train_labels, test_labels = train_test_split(
    my_df['Description'], my_df['Label'], test_size=0.2, random_state=42
)

# Convert text data to numerical features using TF-IDF vectorization
vectorizer = TfidfVectorizer()
train_features = vectorizer.fit_transform(train_data)
test_features = vectorizer.transform(test_data)

# Initialize SVM classifier (you may need to tune hyperparameters)
svm_classifier = SVC(kernel='linear', C=1.0)

# Train the SVM classifier
svm_classifier.fit(train_features, train_labels)

# Predict on the test set
predictions = svm_classifier.predict(test_features)

# Evaluate the performance
accuracy_svm = accuracy_score(test_labels, predictions)
report_svm = classification_report(test_labels, predictions)

print(f"Accuracy: {accuracy_svm:.2f}")
print("\nClassification Report:\n", report_svm)
```
